```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
spine = pd.read_csv("spine.csv")
```

```python
spine.info()
```

## Getting a Big Picture of the Data

```python
spine.describe()
```

### It is important to note the mean is consistently greater than the median, indicating skew

## Looking at Pain

```python
spine["pain"].value_counts()
```

```python
spine["pain"].value_counts().plot.bar()
```

### There seems to be a lot more people with pain than without

```python
spine.groupby("pain").mean()
```

### At an eyeball-glance, the following attributes have large differences:
#### pelvic_incidence
#### lumbar_lordosis_angle
#### sacral_slope
#### pelvic_radius
#### degree_spondylolisthesis

## Looking at Outliers for Each Variable

```python
fig, axes = plt.subplots(3, 4, figsize = (15,20))
axes = axes.flatten()

for i in range(0,len(spine.columns)-1):
    sns.boxenplot(x="pain", y=spine.iloc[:,i], data=spine, orient='v', ax=axes[i],palette="magma_r")

plt.tight_layout()
plt.show()
```

## Looking at Correlations Between Variables

```python
plt.subplots(figsize=(12,8))
sns.heatmap(spine.corr(), annot=True)
```

### It is important to note the variables that are highly correlated with one another

## Looking at Variables Distribution

```python
spine.hist(figsize=(15,12),bins = 20, color="darkmagenta")
plt.title("Features Distribution")
plt.show()
```

### It is interesting to note that the variables that were highly correlated with one another
### also seem to be mostly either normally distributed or positively skewed.
